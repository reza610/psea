```python
import pandas as pd
```

```python
indir = "../testdata/"
outdir = "../../outpsea/"
```

```python
min_mean_expression = 0.1
min_comorbids_percent=0.05
max_comorbids_percent=0.95
samplename="Patient"
total_samples=254
```

```python
expression_df = pd.read_csv(indir+"value_expression.csv", index_col=0)
expression_df_nosamples = expression_df.drop(columns=[samplename])
```

```python
meansdf=expression_df_nosamples.mean(axis=0).to_frame()
meansdf.columns = ["mean_value"]
meansdf["valuename"]= meansdf.index
```

```python
meansdf_include = meansdf[meansdf["mean_value"]>min_mean_expression]
meansdf_include
```

```python
meansdf_include[["valuename"]].to_csv(outdir+"include_values_long.csv", header=False, index=False)
```

```python
badf = pd.read_csv(indir+"comorbid_file.csv", index_col=0)
```

```python
badf
```

```python
badf_noPatient = badf.drop(columns=["Patient"])
badf_noPatient
```

```python
badf_noPatient = badf.drop(columns=["Patient"])
badf_noPatient
```

```python
n_comorbid = badf_noPatient.sum(axis=0).to_frame()
n_comorbid.columns =["n"]
n_comorbid["binary_attribute"] = n_comorbid.index
n_comorbid["percent"] =  n_comorbid["n"]/total_samples
n_comorbid
```

```python
n_comorbid = n_comorbid[n_comorbid["percent"]<max_comorbids_percent]
n_comorbid = n_comorbid[n_comorbid["percent"]>min_comorbids_percent]
n_comorbid
```

```python
n_comorbid[["binary_attribute"]].to_csv(outdir+"include_binary_attribute_long.csv", header=False, index=False)
```
