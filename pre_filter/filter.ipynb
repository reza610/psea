```python
# get working dir and read in the CSV file
os.getcwd()
file_path = '../testdata/comorbid_file.csv'
print(file_path)
df = pd.read_csv(file_path)
```

```python
# Exclude the first 2 columns and convert the rest to numeric, coerce coverts invalid strings to NaN
df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

# Handle any NaN values by converting to 0
df.fillna(0, inplace=True) 

# Count the occurrences of each comorbidity
comorbidity_counts = df.iloc[:, 2:].sum()
print(comorbidity_counts)
```

```python
# Filter out comorbidities that occur in fewer than 2 patients
total_comorbid_threshold = 2
filtered_comorbidities = comorbidity_counts[comorbidity_counts >= total_comorbid_threshold].index
```

```python
# Filter the original DataFrame to keep only the relevant comorbidities
filtered_df = df[['Patient'] + list(filtered_comorbidities)]
print(filtered_df)
```

```python
# filter out pateients with 0 comorbidities
patient_comorbid_threshold = 0
filtered_df = filtered_df[filtered_df.iloc[:, 1:].sum(axis=1) > patient_comorbid_threshold]
print(filtered_df)
```

```python
# get working dir and read in the CSV file for gene expression data
os.getcwd()
file_path = '../testdata/value_expression.csv'
print(file_path)
# Drop the first unnamed column, move the last column to first
df = pd.read_csv(file_path)
df = df.drop(df.columns[0], axis=1)
df = df[[df.columns[-1]] + list(df.columns[:-1])]
print(df)
```

```python
# convert values (except first column) to numeric
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
print(df)
```

```python
# Define your thresholds for gene expression
gene_expression_threshold = 10 # expression of gene is low for everyone
individual_expression_threshold = 10 #individual has overall low expression across genes
```

```python
# Remove genes where all individuals have expression below the threshold
gene_mask = (df.iloc[:, 1:] >= gene_expression_threshold).any(axis=0)
print(gene_mask)
# Ensure that the mask aligns with the DataFrame columns
if len(gene_mask) != df.shape[1] - 1:
    raise ValueError("The length of gene_mask does not match the number of gene columns in df.")
```

```python
# Get the column names corresponding to True values
gene_columns_to_keep = gene_mask.index[gene_mask].tolist()
df_filtered_genes = df.loc[:, [df.columns[0]] + gene_columns_to_keep]
print(df_filtered_genes)
```

```python
# Remove individuals where all gene expression values are below the threshold
individual_mask = df_filtered_genes.iloc[:, 1:].apply(lambda row: (row >= individual_expression_threshold).any(), axis=1)
print(individual_mask)
```

```python
# Filter the DataFrame using the individual_mask
df_filtered = df_filtered_genes[individual_mask]
print(df_filtered)
```

```python
# get working dir and read in the CSV file
sample_name = 'Patient'
binary_attribute_file = '../testdata/comorbid_file.csv'
file_path = binary_attribute_file
print(file_path)
df = pd.read_csv(file_path, index_col=0)
# number samples
total_samples = df.shape[0]
df
```

```python
# Filter out patients without comorbidities
patient_comorbid_threshold = 1
filtered_df = df[df.drop(columns=[sample_name]).sum(axis=1) >= patient_comorbid_threshold]
filtered_df
```

```python
# drop sample_names column
df_noname = filtered_df.drop(columns=[sample_name])
df_noname
```

```python
# Calculate percent occurence of comorbidities
n_comorbid = df_noname.sum(axis=0).to_frame()
n_comorbid.columns =["n"]
n_comorbid["binary_attribute"] = n_comorbid.index
n_comorbid["percent"] =  n_comorbid["n"]/total_samples
n_comorbid
```

```python
# Filter based on user input min and max comorbid occurence
max_comorbids_percent = 0.9
min_comorbids_percent = 0.1
n_comorbid = n_comorbid[n_comorbid["percent"]<max_comorbids_percent]
n_comorbid = n_comorbid[n_comorbid["percent"]>min_comorbids_percent]
n_comorbid
```

```python
#name the output file and and save it
include_binary_attribute_file = '../testdata/include_binary_attribute_long.csv'
outdir = include_binary_attribute_file
n_comorbid[["binary_attribute"]].to_csv(outdir, header=False, index=False)
```

```python
## now do gene expression ##
# get working dir and read in the CSV file for gene expression data
values_file = '../testdata/value_expression.csv'
file_path = values_file
df = pd.read_csv(file_path, index_col=0)
print(file_path)
df
```

```python
# drop samples column
expression_df_nosamples = df.drop(columns=[sample_name])
expression_df_nosamples
```

```python
# Remove individuals (rows) where all gene expression values are below the threshold
individual_expression_threshold = 10
individual_mask = expression_df_nosamples.apply(lambda row: (row >= individual_expression_threshold).any(), axis=1)
individual_mask
# Filter the DataFrame using the individual_mask
expression_df_nosamples = expression_df_nosamples[individual_mask]
expression_df_nosamples
```

```python
# get mean expression, turn into pandas dataframe
min_mean_expression = 0.1
meansdf=expression_df_nosamples.mean(axis=0).to_frame()
meansdf.columns = ["mean_value"]
meansdf["valuename"]= meansdf.index
meansdf
```

```python
# filter based on mean expression
meansdf_include = meansdf[meansdf["mean_value"]>min_mean_expression]
meansdf_include
```

```python
include_values_file = "../testdata/include_values_long.csv"
outdir = include_values_file
meansdf_include[["valuename"]].to_csv(oudir, header=False, index=False)

#looks good
#update the filter script
```
